```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('alert_data.csv')
df
```

```python
df.info()
```

```python
# quitar los [] de las columnas
df['roll'] = df['roll'].str.strip('[]')
df['roll'] = df['roll'].astype(float)
df['pitch'] = df['pitch'].str.strip('[]')
df['pitch'] = df['pitch'].astype(float)
df['yaw'] = df['yaw'].str.strip('[]')
df['yaw'] = df['yaw'].astype(float)
```

```python
df['distracted'].value_counts()
```

```python
# plot time series of the data
plt.figure(figsize=(15, 6))
plt.plot(df.index, df['distracted'])
```

```python
# plot time series of the data
plt.figure(figsize=(15, 6))
plt.plot(df.index, df['ear'])
plt.plot(df.index, df['distracted'])
```

```python
plt.figure(figsize=(15, 6))
plt.plot(df.index, df['roll'])
```

```python
# Plotting the data
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(x='ear', data=df, palette='viridis')
plt.title('Alert Type Distribution')
```
